## Cantidades medibles

* Brillo
    * Los astronomos utilizan la medida llamada magnitud, introducida por Hipparcos cientos de años a. C.

* Color
    * Los astronomos utilizan el color de las estrellas porque se relaciona con su temperatura y tipo espectral

* Posicion
    * Utilizamos las coordenadas celestes de cada astro para conocer a que region pertenecen y si estan cerca de otras estrellas
   

* Distancia
    * Es muy importante medir la distancia a cada estrella, aunque es rara vez posible de hacer.
