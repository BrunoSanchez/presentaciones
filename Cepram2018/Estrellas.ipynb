## Cantidades medibles

* Posicion
    * Utilizamos las coordenadas celestes de cada astro para conocer a que region pertenecen y si estan cerca de otras estrellas
   

* Brillo
    * Los astronomos utilizan la medida llamada magnitud, introducida por Hipparcos cientos de años a. C.

* Color
    * Los astronomos utilizan el color de las estrellas porque se relaciona con su temperatura y tipo espectral

## Cantidades medibles

* Distancia
    * Es muy importante medir la distancia a cada estrella, aunque es rara vez posible de hacer. Se utilizan metodos directos e indirectos.

* Metalicidad
    * La cantidad de elementos quimicos diferentes que la estrella posee. Es posible debido a las lineas de emision de cada elemento de la tabla periodica de elementos.

## Cantidades medibles

* Tamaño de la estrella
    * Usando la temperatura, se puede saber que tipo espectral es cada estrella. Ademas, usando la distancia se puede saber su luminosidad. Y las estrellas mas grandes son mas luminosas.

* Movimiento
    * Se puede utilizar el efecto Doppler para calcular su velocidad radial.
    * Se puede observar durante mucho tiempo para verlas moverse.
